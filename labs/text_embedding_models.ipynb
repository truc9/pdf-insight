```python
from langchain_community.embeddings import OllamaEmbeddings
import numpy as np

oe = OllamaEmbeddings()

embeddings = oe.embed_documents(
    [
        "Hi there!",
        "Oh, Hello",
        "Fake sake",
        "What is your name ?",
        "Yo yo"
    ]
)

v = embeddings[1]

out = np.reshape(v, (len(v), 1))
print(out)

query = oe.embed_query("Hey")
print(query)
```
